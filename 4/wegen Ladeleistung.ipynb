- um die Kalkulierung zu vereinfachen, kann hinzugefügt werden, dass die Ladeleistung (charging power) relativ gleichbleibend und für die allermeisten sessions in einem engen Korridor liegt (siehe unten)
    - site 1: mean 3,5 kW
    - site 2: mean 4,4 kW

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('../data/processed/charging_sessions_cleaned.csv')
df['connectionTime'] = pd.to_datetime(df['connectionTime'])
df['disconnectTime'] = pd.to_datetime(df['disconnectTime'])
df = df.loc[df['chargingPower'] < 100]
```

```python
df1 = df.loc[df['siteID'] == 1]
df2 = df.loc[df['siteID'] == 2]
```

```python
df1['chargingPower'].describe()
```

```python
df2['chargingPower'].describe()
```

```python
#plot chargingpower for site 1
plt.figure(figsize=(30, 6))
plt.scatter(df1['connectionTime'], df1["chargingPower"], s=.2, alpha=0.7)
plt.xlabel("time")
plt.ylabel("Charging Power")
plt.title("site 1: charging power")
plt.ylim(0,20)
plt.show()
```

```python
#plot chargingpower for site 2
plt.figure(figsize=(30, 6))
plt.scatter(df2['connectionTime'], df2["chargingPower"], s=.2, alpha=0.7)
plt.xlabel("time")
plt.ylabel("Charging Power")
plt.title("site 2: charging power")
plt.ylim(0,20)
plt.show()
```
